```python
folder = './data/uf_mix'
cursor = conn.cursor()

for num, gid in enumerate(ids[0:5]):
    print(gid[0]," - ",num,"/",len(ids))
    sql = "select problem from ga_run where id = {};".format(gid[0])
    problem = sqlio.read_sql_query(sql, conn)
    num_vars, clauses = fn.read_problem(folder+'/'+problem.values[0][0])
    sql = "select id, population from ga_run_population where ga_run_id = {} order by id asc;".format(gid[0])
    dat = sqlio.read_sql_query(sql, conn)
    
    pop_fitness_dict = {}
    
    for i in range(len(dat)):
        if i%100==0: print("Generation: ",(i+1))
        population = dat['population'].values[i]

        pop_fitness, max_fitness, max_fit_indiv = fn.evaluate_population(population, clauses, fn.maxsat_fitness, pop_fitness_dict, 1)
        fitness_arr = []            
        for indiv, fitness in pop_fitness:
            fitness_arr += [fitness]
            if type(indiv) == type(()):
                pop_fitness_dict[indiv]=fitness
            else:
                pop_fitness_dict[tuple(indiv)]=fitness

        mean = float(numpy.mean(fitness_arr))
        median = float(numpy.median(fitness_arr))
        mode = stats.mode(fitness_arr)
        mode_value = float(mode[0][0])
        mode_count = int(mode[1][0])
        deviation = float(numpy.std(fitness_arr))
        maximum = max(fitness_arr)
        
        if i%100==0: print ("update ga_run_generations set mean=",mean,", median=",median,", mode=",mode_value,", mode_count=",mode_count,", standard_deviation=",deviation," WHERE ga_run_id=",gid[0]," and generation_num=",i+1)
        cursor.execute("update ga_run_generations set mean=(%s), median=(%s), mode=(%s), mode_count=(%s), standard_deviation=(%s) WHERE ga_run_id=(%s) and generation_num=(%s)",(mean, median, mode_value, mode_count, deviation, gid[0], i+1,))
        conn.commit()
        
cursor.close()
```
